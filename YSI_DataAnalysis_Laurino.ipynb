```python
#Import necessary libraries:
import pandas as pd #Data import
import numpy as np #Math functions
import matplotlib.pyplot as plt #Graphing library
```

```python
#Import data using Pandas:
data = pd.read_excel('JanetFaccenda_DataSheet_2018.xlsx')
#View the data
data
```

```python
#List the column names:
print(data.columns)
```

```python
#Simplify the column headers for easier data processing:
data = data.rename(columns={'Tide Value' : 'TideValue',
                            'pH Temperature (ËšC)': 'pHTempC',
                            'pH Temperature (ËšF)': 'pHTempF',
                            'Date (MM/DD/YY)':'DATE',
                            'Multimeter Water Temperature': 'MeterTemp',
                            'Water Temperature' : 'WaterTemp',
                            'Dissolved O2 (mg/L)' : 'DissO2mgL',
                            'Dissolved O2 (%)' : 'DissO2%',
                            'Barometric Pressure (mmHg)' : 'BaroPressMMHG',
                            'Specific Conductivity (mS/cm)':'SpecificConductivity',
                            'Conductivity (mS/cm)':'Conductivity',
                            'Salinity (ppt)':'Salinity',
                           })
```

```python
#Remove all spaces in the other column names and list the new column names:
data.columns = [c.replace(' ', '_') for c in data.columns]
print(data.columns)
```

```python
plt.figure(figsize=(20,8))
plt.plot(data.DATE, data.pHTempC, "b-", label='C',linewidth= 4,markersize=5, alpha=.80)
plt.plot(data.DATE, data.pHTempF, "r-", label='F',linewidth= 4,markersize=5, alpha=.80)
plt.ylabel('Temperature', fontsize=16)
plt.xlabel('Date', fontsize=16)
plt.legend()
plt.title('Multimeter Water Temperature', fontsize=20)
plt.show()
```

```python
plt.figure(figsize=(20,8))
plt.plot(data.Latitude, data.Salinity, "b.", label='',linewidth= 4,markersize=15, alpha=.80)
#plt.plot(data.Longitude, data.Salinity, "r.", label='F',linewidth= 4,markersize=5, alpha=.80)
plt.ylabel('Salinity (PPT)', fontsize=16)
plt.xlabel('Latitude', fontsize=16)
plt.legend()
plt.title('Multimeter Salinity', fontsize=20)
plt.show()
```

```python
plt.figure(figsize=(20,8))
#plt.plot(data.Latitude, data.Salinity, "b.", label='',linewidth= 4,markersize=15, alpha=.80)
plt.plot(data.Longitude, data.Salinity, "r.", label='',linewidth= 4,markersize=15, alpha=.80)
plt.ylabel('Salinity (PPT)', fontsize=16)
plt.xlabel('Longitude', fontsize=16)
plt.legend()
plt.title('Multimeter Salinity', fontsize=20)
plt.show()
```

```python
plt.figure(figsize=(20,8))
plt.plot(data.MeterTemp, data.Salinity, "b.", label='Salinity',linewidth= 4,markersize=15, alpha=.80)
plt.plot(data.MeterTemp, data.Depth, "r.", label='Depth',linewidth= 4,markersize=15, alpha=.80)
plt.ylabel('Salinity (PPT)', fontsize=16)
plt.xlabel('Water Temperature', fontsize=16)
plt.legend()
plt.title('Multimeter Salinity, Depth and Water Temperature', fontsize=20)
plt.show()
```

```python
plt.figure(figsize=(20,8))
plt.plot(data.DATE, data.Salinity, "b.", label='Salinity (PPT)',linewidth= 4,markersize=15, alpha=.80)
plt.plot(data.DATE, data.Depth, "r.", label='Depth (FT)',linewidth= 4,markersize=15, alpha=.80)
plt.plot(data.DATE, data.MeterTemp, "g.", label='Meter Temperature (F)',linewidth= 4,markersize=15, alpha=.80)
plt.ylabel(' ', fontsize=16)
plt.xlabel('Date', fontsize=16)
plt.legend()
plt.title('Multimeter Salinity, Depth and Water Temperature', fontsize=20)
plt.show()
```

```python
print("Number of recordings per data value:")

data['Atmosphere'].value_counts()[:30].plot(kind='barh')
plt.title('Atmosphere Recordings', size=15)
plt.show()

data['Precipitation'].value_counts()[:30].plot(kind='barh')
plt.title('Precipitation Recordings', size=15)
plt.show()

data['Wind_Direction'].value_counts()[:30].plot(kind='barh')
plt.title('Wind Direction Recordings', size=15)
plt.show()

data['Wind_Speed'].value_counts()[:30].plot(kind='barh')
plt.title('Wind Speed Recordings', size=15)
plt.show()

data['Swell_Direction'].value_counts()[:30].plot(kind='barh')
plt.title('Swell Direction Recordings', size=15)
plt.show()

data['Swell_Height'].value_counts()[:30].plot(kind='barh')
plt.title('Swell Height Recordings', size=15)
plt.show()

data['Area'].value_counts()[:30].plot(kind='barh')
plt.title('Area Recordings', size=15)
plt.show()

data['Tide_Stage'].value_counts()[:30].plot(kind='barh')
plt.title('Tide Recordings', size=15)
plt.show()
```
